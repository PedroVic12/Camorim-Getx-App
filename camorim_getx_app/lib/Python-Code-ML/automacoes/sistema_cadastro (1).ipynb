```python
def escreve_formulario(x,y,dados):
    pyautogui.click(x,y, duration = 0.5)
    pyautogui.write(dados)

def get_last_row(planilha):
    return planilha.max_row

```

```python


workbook = openpyxl.load_workbook(PLANILHA)
planilha = workbook['OverView IFS']
_array = []

lastRow = get_last_row(planilha)

for row in planilha.iter_rows(min_row = 2,max_row=6):

    # COluna1 , Coluna 2, Coluna 3
    equipamento = row[1].value
    print(equipamento)
    escreve_formulario(1149,178,equipamento)

    rebocador = row[0].value
    escreve_formulario(1150,246,rebocador)

    acao = row[2].value
    escreve_formulario(1150,323,acao)

    status = row[3].value
    escreve_formulario(1150,370,status)

    tipo = row[4].value
    escreve_formulario(1160,467,tipo)

    pyautogui.click(1319,724)

    
```

```python

data_map ={}


try:
    for i in range(0,lastRow):
        dados = row[i].value
        print(dados)

        print(f'linha {i}' ,' = ',dados, end=' | ')
        data_map['linha'] = i
        data_map['dados'] = dados
        _array.append(data_map)

except:
    print('erro')



for item in _array:
    print(item)
```
