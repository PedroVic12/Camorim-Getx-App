```python
import openpyxl
import pyautogui
import pandas as pd
PLANILHA = '../assets/EQUIPAMENTOS FROTA.xlsx'

def escreve_formulario(x,y,dados):
    pyautogui.click(x,y, duration = 0.5)
    pyautogui.write(dados)

def get_last_row(planilha):
    return planilha.max_row

```

```python
df_equipamentos = pd.read_excel(PLANILHA)
cols = df_equipamentos.columns[:4]

df = df_equipamentos[cols]
df
```

```python
df_cadastro = pd.read_excel('../assets/Cadastro de Equipamentos - Oilxplorer Marine.xlsx',skiprows=6)
df_cadastro.head()
```

```python
# Carregar a planilha
workbook = openpyxl.load_workbook('../assets/Cadastro de Equipamentos - Oilxplorer Marine.xlsx')
sheet = workbook['Plan1']  # Substitua pelo nome real da sua aba

# Começar a escrever a partir da linha 9
start_row = 9
for index, row in df.iterrows():
    # Escrevendo os dados nas células correspondentes
    sheet[f'B{start_row + index}'] = row['EQUIPAMENTO']
    sheet[f'C{start_row + index}'] = row['EMBARCAÇÃO']
    sheet[f'D{start_row + index}'] = row['FABRICANTE']
    sheet[f'E{start_row + index}'] = row['MODELO']
    # Adicione mais células conforme necessário

# Salvar a planilha
workbook.save('output.xlsx')
```

```python
# Mapear as informações do df1 para o df2
for index, row in df_equipamentos.iterrows():
    # Encontre a linha correspondente em df2 (ajuste a condição conforme necessário)
    correspondente = df_cadastro[df_cadastro['Descrição do Equipamento'].str.contains(row['EQUIPAMENTO'], na=False)]
    if not correspondente.empty:
        df2_index = correspondente.index[0]
        df_cadastro.at[df2_index, 'Lubrificante em uso'] = 'O lubrificante que deseja'
        df_cadastro.at[df2_index, 'Fabricante do Equipamento'] = row['FABRICANTE']
        df_cadastro.at[df2_index, 'Modelo do Equipamento'] = row['MODELO']
        # Continue com outros campos conforme necessário

# Salvar a planilha atualizada
with pd.ExcelWriter('caminho_para_planilha.xlsx', engine='openpyxl') as writer:
    df_cadastro.to_excel(writer, sheet_name='Nome_da_Aba', index=False)
```
